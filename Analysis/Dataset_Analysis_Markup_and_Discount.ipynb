```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
markups = pd.read_csv('StarMart_Markup_Discount.csv')
markups.head()
```

```python
markups.describe().round(2)
```

```python
sns.histplot(markups['markup'], bins=30)
plt.ylabel('Count')
plt.xlabel('Markup Distribution')
plt.title("Distribution of Product Markups")

plt.tight_layout()
plt.show()
```

```python
markups['normal_day_flag'] = np.where(markups['normal_day_discount'] == 0, False, True)
```

```python
sns.countplot(data=markups, x='normal_day_flag')
plt.title('Count of Normal Day Flag')
plt.xlabel('Normal Day Flag')
plt.ylabel('Count')
plt.show()
```

```python
markups['holiday_flag'] = np.where(markups['holiday_discount'] == 0, False, True)
```

```python
sns.countplot(data=markups, x='holiday_flag')
plt.title('Count of Normal Day Flag')
plt.xlabel('Normal Day Flag')
plt.ylabel('Count')
plt.show()
```

```python
sns.kdeplot(markups['normal_day_discount'], label='Normal Day', color='lightcoral')
sns.kdeplot(markups['holiday_discount'], label='HoliDay', color='lightgreen')
plt.ylabel('Count')
plt.xlabel('Markup Distribution')
plt.title("Distribution of Product Markups")
plt.legend()

plt.tight_layout()
plt.show()
```
