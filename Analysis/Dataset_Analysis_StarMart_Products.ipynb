```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Customers 
df = pd.read_csv("StarMart_Products.csv")
df.head()
```

```python
df.describe().round(2)
```

```python
df.describe(include='object')
```

# Product Per Store

```python
plt.figure(figsize=(15, 10))
sns.countplot(data=df, y='store_id', order=df["store_id"].value_counts().index, palette="viridis", hue='store_id')
plt.title("Product Count per Store", fontsize=16, weight='bold')
plt.xlabel("Number of Products", fontsize=14)
plt.ylabel("Stores", fontsize=14)
plt.tight_layout()
plt.show()
```

# Number of Products per Category

```python
plt.figure(figsize=(15, 10))
sns.countplot(data=df, y='category', order=df["category"].value_counts().index, palette="viridis", hue='category')
plt.title("Product Count per Category", fontsize=16, weight='bold')
plt.xlabel("Number of Products", fontsize=14)
plt.ylabel("Category", fontsize=14)
plt.tight_layout()
plt.show()
```

```python
# character limit for SQL

object_cols = df.select_dtypes(include='object')

# Find value with max length in each column
max_length_values = {
    col: object_cols[col].loc[object_cols[col].str.len().idxmax()]
    for col in object_cols.columns
}

# Display results
for col, val in max_length_values.items():
    print(f"Column: {col}\nLength: {len(val)}\nValue: {val}\n{'-'*70}")
```
